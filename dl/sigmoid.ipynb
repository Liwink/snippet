```python
import math
import numpy as np
```

```python
def sigmoid(x):
    return 1 / (1 + np.exp(-x))

def sigmoid_cross_entropy(logits, labels):
    return logits - logits * labels + np.log(1 + np.exp(-logits))
```

```python
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
logits = np.arange(-10, 11)
pos_labels = np.ones(len(logits))
neg_labels = np.zeros(len(logits))
```

```python
plt.plot(logits, sigmoid_cross_entropy(logits, pos_labels))
```

```python
plt.plot(logits, sigmoid_cross_entropy(logits, neg_labels))
```

```python
def sigmoid_cross_entropy_tf(logits, labels):
    return np.maximum(logits, 0) - logits * labels + np.log(1 + np.exp(-np.abs(logits)))
```

```python
plt.plot(logits, sigmoid_cross_entropy_tf(logits, pos_labels))
```

```python
plt.plot(logits, sigmoid_cross_entropy_tf(logits, neg_labels))
```
