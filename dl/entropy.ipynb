```python
import matplotlib.pyplot as plt
import numpy as np

%matplotlib inline
```

```python
x = np.arange(0.01, 1, 0.01)
```

```python
y_log = np.log(x)
```

```python
plt.plot(x, y_log)
```

```python
def entropy(y_hat, y):
    return -(y * np.log(y_hat) + (1 - y) * np.log(1 - y_hat))
```

```python
plt.plot(x, entropy(x, 1))
```

```python
plt.plot(x, entropy(x, 0))
```
